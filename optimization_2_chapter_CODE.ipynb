```python
# Оптимизация весов метрик для оценки сходства программного кода
# Сравнение 4 алгоритмов: GA, PSO, DE, NRO
# Метрики: Хэлстед, Структурные метрики, Токен-based метрики

import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from mealpy import FloatVar
from mealpy.evolutionary_based import GA, DE
from mealpy.swarm_based import PSO
from mealpy.physics_based import NRO
import warnings
import logging
import matplotlib.pyplot as plt
import seaborn as sns
import time
import re
import ast
from collections import Counter

# Отключение всех логов
warnings.filterwarnings('ignore')
logging.getLogger('mealpy').setLevel(logging.ERROR)

# ==============================================================================
# 1. ЗАГРУЗКА И ПОДГОТОВКА ДАННЫХ
# ==============================================================================

print("="*80)
print("ЗАГРУЗКА ДАННЫХ")
print("="*80)

# Загрузка данных
expert_matrix_df = pd.read_excel('expert_matrix_code.xlsx', header=0, index_col=0)
codes_df = pd.read_excel('selected_codes.xlsx')

expert_matrix = expert_matrix_df.values
n_codes = len(codes_df)

print(f"✓ Загружено фрагментов кода: {n_codes}")
print(f"✓ Размер экспертной матрицы: {expert_matrix.shape}")

# Извлечение кода
codes = codes_df['code'].fillna('').tolist()

print(f"✓ Код подготовлен к анализу")

# ==============================================================================
# 2. МЕТРИКА 1: МЕТРИКИ ХЭЛСТЕДА
# ==============================================================================

print("\n" + "="*80)
print("ВЫЧИСЛЕНИЕ МЕТРИК КОДА")
print("="*80)

print("\n[1/3] Вычисление метрик Хэлстеда...")

def extract_operators_operands(code):
    """Извлечение операторов и операндов из кода Python"""
    operators = set()
    operands = set()
    
    # Python операторы
    python_operators = {
        '+', '-', '*', '/', '//', '%', '**',
        '=', '+=', '-=', '*=', '/=', '//=', '%=', '**=',
        '==', '!=', '<', '>', '<=', '>=',
        'and', 'or', 'not', 'is', 'in',
        '&', '|', '^', '~', '<<', '>>',
        'if', 'else', 'elif', 'for', 'while', 'break', 'continue',
        'def', 'class', 'return', 'yield', 'import', 'from', 'as',
        'try', 'except', 'finally', 'raise', 'with',
        'lambda', 'pass', 'del', 'global', 'nonlocal'
    }
    
    # Подсчет операторов и операндов
    tokens = re.findall(r'\b\w+\b|[+\-*/%=<>!&|^~]+|\(|\)|\[|\]|\{|\}|,|:|;|\.', code)
    
    for token in tokens:
        if token in python_operators or token in '()[]{},:;.':
            operators.add(token)
        elif token and not token.isspace():
            operands.add(token)
    
    return operators, operands

def calculate_halstead_metrics(code):
    """Вычисление метрик Хэлстеда"""
    try:
        operators, operands = extract_operators_operands(code)
        
        n1 = len(operators)  # Число уникальных операторов
        n2 = len(operands)   # Число уникальных операндов
        
        # Подсчет общего числа операторов и операндов
        tokens = re.findall(r'\b\w+\b|[+\-*/%=<>!&|^~]+|\(|\)|\[|\]|\{|\}|,|:|;|\.', code)
        N1 = sum(1 for t in tokens if t in operators or t in '()[]{},:;.')
        N2 = len([t for t in tokens if t not in operators and t not in '()[]{},:;.' and t.strip()])
        
        # Метрики Хэлстеда
        vocabulary = n1 + n2
        length = N1 + N2
        
        if vocabulary == 0:
            return [0, 0, 0, 0, 0]
        
        calculated_length = n1 * np.log2(n1 + 1) + n2 * np.log2(n2 + 1) if n1 > 0 or n2 > 0 else 0
        volume = length * np.log2(vocabulary) if vocabulary > 1 else 0
        difficulty = (n1 / 2) * (N2 / (n2 + 1)) if n2 > 0 else 0
        effort = difficulty * volume
        
        return [vocabulary, length, volume, difficulty, effort]
    except:
        return [0, 0, 0, 0, 0]

halstead_features = []
for code in codes:
    halstead_features.append(calculate_halstead_metrics(code))

halstead_matrix = np.array(halstead_features)
print(f"✓ Метрики Хэлстеда вычислены: {halstead_matrix.shape}")

# Нормализация и создание матрицы сходства
from sklearn.preprocessing import StandardScaler
scaler_halstead = StandardScaler()
halstead_normalized = scaler_halstead.fit_transform(halstead_matrix)
halstead_similarity = cosine_similarity(halstead_normalized)

# ==============================================================================
# 3. МЕТРИКА 2: СТРУКТУРНЫЕ МЕТРИКИ
# ==============================================================================

print("\n[2/3] Вычисление структурных метрик...")

def calculate_structural_metrics(code):
    """Вычисление структурных метрик кода"""
    try:
        # Базовые метрики
        loc = len([line for line in code.split('\n') if line.strip()])  # Строки кода
        total_chars = len(code)
        avg_line_length = total_chars / loc if loc > 0 else 0
        
        # Глубина вложенности
        max_indent = 0
        for line in code.split('\n'):
            if line.strip():
                indent = len(line) - len(line.lstrip())
                max_indent = max(max_indent, indent // 4)  # Предполагаем 4 пробела
        
        # Количество функций и классов
        num_functions = code.count('def ')
        num_classes = code.count('class ')
        
        # Количество циклов и условий
        num_loops = code.count('for ') + code.count('while ')
        num_conditions = code.count('if ') + code.count('elif ')
        
        # Количество импортов
        num_imports = code.count('import ') + code.count('from ')
        
        # Количество комментариев
        num_comments = code.count('#')
        
        return [loc, total_chars, avg_line_length, max_indent, 
                num_functions, num_classes, num_loops, num_conditions,
                num_imports, num_comments]
    except:
        return [0] * 10

structural_features = []
for code in codes:
    structural_features.append(calculate_structural_metrics(code))

structural_matrix = np.array(structural_features)
print(f"✓ Структурные метрики вычислены: {structural_matrix.shape}")

# Нормализация и создание матрицы сходства
scaler_structural = StandardScaler()
structural_normalized = scaler_structural.fit_transform(structural_matrix)
structural_similarity = cosine_similarity(structural_normalized)

# ==============================================================================
# 4. МЕТРИКА 3: ТОКЕН-BASED МЕТРИКИ
# ==============================================================================

print("\n[3/3] Вычисление токен-based метрик...")

def tokenize_code(code):
    """Токенизация кода Python"""
    try:
        # Удаление комментариев
        code_no_comments = re.sub(r'#.*', '', code)
        
        # Извлечение токенов (идентификаторы, ключевые слова, операторы)
        tokens = re.findall(r'\b[a-zA-Z_][a-zA-Z0-9_]*\b', code_no_comments)
        
        # Фильтрация служебных слов Python
        python_keywords = {
            'False', 'None', 'True', 'and', 'as', 'assert', 'async', 'await',
            'break', 'class', 'continue', 'def', 'del', 'elif', 'else', 'except',
            'finally', 'for', 'from', 'global', 'if', 'import', 'in', 'is',
            'lambda', 'nonlocal', 'not', 'or', 'pass', 'raise', 'return',
            'try', 'while', 'with', 'yield'
        }
        
        # Оставляем все токены (включая ключевые слова, так как они важны для структуры)
        return ' '.join(tokens)
    except:
        return ''

tokenized_codes = [tokenize_code(code) for code in codes]

# TF-IDF векторизация токенов
tfidf_vectorizer = TfidfVectorizer(max_features=200, min_df=1, ngram_range=(1, 2))
token_matrix = tfidf_vectorizer.fit_transform(tokenized_codes).toarray()
token_similarity = cosine_similarity(token_matrix)

print(f"✓ Токен-based метрики вычислены: {token_matrix.shape}")

# ==============================================================================
# 5. НОРМАЛИЗАЦИЯ МАТРИЦ СХОДСТВА
# ==============================================================================

def normalize_similarity_matrix(matrix):
    """Нормализация матрицы сходства к диапазону [0, 1]"""
    matrix_norm = matrix.copy()
    min_val, max_val = matrix_norm.min(), matrix_norm.max()
    
    if max_val > min_val:
        matrix_norm = (matrix_norm - min_val) / (max_val - min_val)
    
    np.fill_diagonal(matrix_norm, 1.0)
    return matrix_norm

halstead_similarity_norm = normalize_similarity_matrix(halstead_similarity)
structural_similarity_norm = normalize_similarity_matrix(structural_similarity)
token_similarity_norm = normalize_similarity_matrix(token_similarity)

print("\n✓ Все матрицы нормализованы")

# ==============================================================================
# 6. ОПРЕДЕЛЕНИЕ ЗАДАЧИ ОПТИМИЗАЦИИ
# ==============================================================================

def create_automatic_matrix(weights):
    """
    Создание автоматической матрицы сходства с заданными весами
    
    Parameters:
    -----------
    weights : array-like, shape (3,)
        Веса для [Halstead, Structural, Token-based]
    
    Returns:
    --------
    weighted_matrix : ndarray
        Взвешенная матрица сходства
    """
    # Нормализация весов
    weights = np.maximum(weights, 0)  # Неотрицательность
    total = weights.sum()
    weights = weights / total if total > 0 else np.ones(3) / 3
    
    # Взвешенная комбинация
    weighted_matrix = (
        weights[0] * halstead_similarity_norm +
        weights[1] * structural_similarity_norm +
        weights[2] * token_similarity_norm
    )
    
    return weighted_matrix

def objective_function(solution):
    """
    Целевая функция для оптимизации: расстояние Фробениуса
    между автоматической и экспертной матрицами
    """
    auto_matrix = create_automatic_matrix(solution)
    
    # Ограничиваем размер для сравнения
    n = min(auto_matrix.shape[0], expert_matrix.shape[0])
    
    # Расстояние Фробениуса
    distance = np.linalg.norm(
        auto_matrix[:n, :n] - expert_matrix[:n, :n],
        ord='fro'
    )
    
    return distance

# ==============================================================================
# 7. НАСТРОЙКА АЛГОРИТМОВ ОПТИМИЗАЦИИ
# ==============================================================================

print("\n" + "="*80)
print("НАСТРОЙКА АЛГОРИТМОВ ОПТИМИЗАЦИИ")
print("="*80)

# Определение границ для весов
bounds = [
    FloatVar(lb=0.0, ub=1.0, name="w_halstead"),
    FloatVar(lb=0.0, ub=1.0, name="w_structural"),
    FloatVar(lb=0.0, ub=1.0, name="w_token")
]

# Создание задачи оптимизации
problem_dict = {
    "obj_func": objective_function,
    "bounds": bounds,
    "minmax": "min"
}

# Общие параметры
EPOCH = 50
POP_SIZE = 30
NUM_RUNS = 10

# Параметры для каждого алгоритма
algorithms_config = {
    'GA': {
        'class': GA.BaseGA,
        'params': {
            'epoch': EPOCH,
            'pop_size': POP_SIZE,
            'pc': 0.9,
            'pm': 0.1,
            'selection': 'tournament',
            'k_way': 0.2,
            'crossover': 'uniform',
            'mutation': 'flip'
        },
        'name': 'Генетический алгоритм'
    },
    'PSO': {
        'class': PSO.OriginalPSO,
        'params': {
            'epoch': EPOCH,
            'pop_size': POP_SIZE,
            'c1': 2.05,
            'c2': 2.05,
            'w_min': 0.4,
            'w_max': 0.9
        },
        'name': 'Рой частиц'
    },
    'DE': {
        'class': DE.OriginalDE,
        'params': {
            'epoch': EPOCH,
            'pop_size': POP_SIZE,
            'wf': 0.8,
            'cr': 0.9,
            'strategy': 0
        },
        'name': 'Дифференциальная эволюция'
    },
    'NRO': {
        'class': NRO.OriginalNRO,
        'params': {
            'epoch': EPOCH,
            'pop_size': POP_SIZE
        },
        'name': 'Ядерная реакция'
    }
}

print("\nНастроенные алгоритмы:")
for algo_name, config in algorithms_config.items():
    print(f"  • {algo_name}: {config['name']}")

# ==============================================================================
# 8. ЗАПУСК ОПТИМИЗАЦИИ ВСЕМИ АЛГОРИТМАМИ
# ==============================================================================

print("\n" + "="*80)
print("МНОЖЕСТВЕННЫЕ ЗАПУСКИ ВСЕХ АЛГОРИТМОВ")
print("="*80)

all_algorithms_results = {}

for algo_name, config in algorithms_config.items():
    print(f"\n{'='*80}")
    print(f"АЛГОРИТМ: {config['name']} ({algo_name})")
    print(f"{'='*80}")
    
    algo_runs_data = []
    
    print(f"\nВыполняется {NUM_RUNS} независимых запусков...")
    
    for run_idx in range(NUM_RUNS):
        print(f"\n[Запуск {run_idx + 1}/{NUM_RUNS}]", end=' ')
        
        # Создание оптимизатора
        optimizer = config['class'](**config['params'])
        
        # Замер времени
        start_time = time.time()
        best_agent = optimizer.solve(problem_dict, mode='single')
        elapsed_time = time.time() - start_time
        
        # Извлечение результатов
        optimal_weights = best_agent.solution
        optimal_weights = np.maximum(optimal_weights, 0)
        optimal_weights = optimal_weights / optimal_weights.sum()
        
        best_distance = best_agent.target.fitness
        
        # Сохранение истории оптимизации
        convergence_curve = optimizer.history.list_global_best_fit
        
        # Сохранение данных запуска
        algo_runs_data.append({
            'run_id': run_idx + 1,
            'weights': optimal_weights.copy(),
            'distance': best_distance,
            'convergence': convergence_curve,
            'time': elapsed_time
        })
        
        print(f"Веса: Halstead={optimal_weights[0]:.4f}, "
              f"Structural={optimal_weights[1]:.4f}, "
              f"Token={optimal_weights[2]:.4f}, "
              f"Dist={best_distance:.6f}, "
              f"Time={elapsed_time:.2f}s")
    
    # Сохранение результатов алгоритма
    all_algorithms_results[algo_name] = {
        'config': config,
        'runs': algo_runs_data
    }
    
    # Статистика по алгоритму
    all_weights = np.array([run['weights'] for run in algo_runs_data])
    all_distances = np.array([run['distance'] for run in algo_runs_data])
    all_times = np.array([run['time'] for run in algo_runs_data])
    
    weights_mean = all_weights.mean(axis=0)
    weights_std = all_weights.std(axis=0)
    
    print(f"\n📊 Статистика {algo_name}:")
    print(f"  Среднее расстояние: {all_distances.mean():.6f} ± {all_distances.std():.6f}")
    print(f"  Лучшее расстояние:  {all_distances.min():.6f}")
    print(f"  Худшее расстояние:  {all_distances.max():.6f}")
    print(f"  Среднее время:      {all_times.mean():.2f}s ± {all_times.std():.2f}s")
    print(f"  Средние веса:       Halstead={weights_mean[0]:.4f}, "
          f"Structural={weights_mean[1]:.4f}, Token={weights_mean[2]:.4f}")

# ==============================================================================
# 9. СРАВНИТЕЛЬНЫЙ АНАЛИЗ АЛГОРИТМОВ
# ==============================================================================

print("\n" + "="*80)
print("СРАВНИТЕЛЬНЫЙ АНАЛИЗ АЛГОРИТМОВ")
print("="*80)

# Таблица сравнения
comparison_table = []
for algo_name, results in all_algorithms_results.items():
    runs = results['runs']
    distances = [run['distance'] for run in runs]
    times = [run['time'] for run in runs]
    weights = np.array([run['weights'] for run in runs])
    
    comparison_table.append({
        'Алгоритм': algo_name,
        'Среднее_расстояние': np.mean(distances),
        'Ст.откл._расстояния': np.std(distances),
        'Лучшее_расстояние': np.min(distances),
        'Худшее_расстояние': np.max(distances),
        'Среднее_время_(с)': np.mean(times),
        'Ст.откл._времени_(с)': np.std(times),
        'Средний_вес_Halstead': weights[:, 0].mean(),
        'Средний_вес_Structural': weights[:, 1].mean(),
        'Средний_вес_Token': weights[:, 2].mean(),
        'CV_Halstead_%': (weights[:, 0].std() / weights[:, 0].mean()) * 100,
        'CV_Structural_%': (weights[:, 1].std() / weights[:, 1].mean()) * 100,
        'CV_Token_%': (weights[:, 2].std() / weights[:, 2].mean()) * 100
    })

comparison_df = pd.DataFrame(comparison_table)
comparison_df = comparison_df.sort_values('Лучшее_расстояние')

print("\n📊 Сводная таблица сравнения алгоритмов:")
print(comparison_df.to_string(index=False))

# Определяем лучший алгоритм
best_algo_name = comparison_df.iloc[0]['Алгоритм']
best_algo_results = all_algorithms_results[best_algo_name]
best_run_idx = np.argmin([run['distance'] for run in best_algo_results['runs']])
best_run = best_algo_results['runs'][best_run_idx]

print(f"\n🏆 ЛУЧШИЙ АЛГОРИТМ: {best_algo_name} ({algorithms_config[best_algo_name]['name']})")
print(f"   • Лучшее расстояние: {best_run['distance']:.6f}")
print(f"   • Оптимальные веса: Halstead={best_run['weights'][0]:.4f}, "
      f"Structural={best_run['weights'][1]:.4f}, Token={best_run['weights'][2]:.4f}")

# ==============================================================================
# 10. ВИЗУАЛИЗАЦИЯ РЕЗУЛЬТАТОВ
# ==============================================================================

print("\n" + "="*80)
print("ВИЗУАЛИЗАЦИЯ РЕЗУЛЬТАТОВ")
print("="*80)

# ==============================================================================
# 10.1. СРАВНЕНИЕ КРИВЫХ СХОДИМОСТИ
# ==============================================================================

fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Кривые сходимости алгоритмов оптимизации', 
             fontsize=16, fontweight='bold')

colors_algo = {
    'GA': '#2ecc71',
    'PSO': '#3498db',
    'DE': '#e74c3c',
    'NRO': '#f39c12'
}

for idx, (algo_name, results) in enumerate(all_algorithms_results.items()):
    ax = axes[idx // 2, idx % 2]
    
    # Рисуем все запуски
    for run in results['runs']:
        conv = run['convergence']
        epochs = range(1, len(conv) + 1)
        ax.plot(epochs, conv, alpha=0.3, color=colors_algo[algo_name], linewidth=1)
    
    # Рисуем среднюю кривую
    all_convergences = np.array([run['convergence'] for run in results['runs']])
    mean_convergence = all_convergences.mean(axis=0)
    epochs = range(1, len(mean_convergence) + 1)
    ax.plot(epochs, mean_convergence, color=colors_algo[algo_name], 
            linewidth=3, label='Среднее', zorder=10)
    
    # Рисуем лучшую кривую
    best_run_idx = np.argmin([run['distance'] for run in results['runs']])
    best_conv = results['runs'][best_run_idx]['convergence']
    epochs = range(1, len(best_conv) + 1)
    ax.plot(epochs, best_conv, color='black', linewidth=2.5, 
            linestyle='--', label='Лучший запуск', zorder=11)
    
    ax.set_xlabel('Поколение', fontsize=11, fontweight='bold')
    ax.set_ylabel('Расстояние Фробениуса', fontsize=11, fontweight='bold')
    ax.set_title(f'{algo_name}: {algorithms_config[algo_name]["name"]}', 
                 fontweight='bold', fontsize=12)
    ax.grid(True, alpha=0.3)
    ax.legend(fontsize=9)

plt.tight_layout()
plt.show()

# ==============================================================================
# 10.2. СРАВНЕНИЕ РАСПРЕДЕЛЕНИЯ ВЕСОВ
# ==============================================================================

fig, axes = plt.subplots(1, 3, figsize=(18, 6))
fig.suptitle('Распределение весов метрик по алгоритмам', 
             fontsize=16, fontweight='bold')

weight_names = ['Halstead', 'Structural', 'Token-based']

for weight_idx, weight_name in enumerate(weight_names):
    ax = axes[weight_idx]
    
    data_for_boxplot = []
    labels_for_boxplot = []
    colors_for_boxplot = []
    
    for algo_name, results in all_algorithms_results.items():
        weights = np.array([run['weights'][weight_idx] for run in results['runs']])
        data_for_boxplot.append(weights)
        labels_for_boxplot.append(algo_name)
        colors_for_boxplot.append(colors_algo[algo_name])
    
    bp = ax.boxplot(data_for_boxplot, labels=labels_for_boxplot, 
                     patch_artist=True, notch=True, showmeans=True)
    
    for patch, color in zip(bp['boxes'], colors_for_boxplot):
        patch.set_facecolor(color)
        patch.set_alpha(0.7)
    
    ax.set_ylabel('Значение веса', fontsize=11, fontweight='bold')
    ax.set_title(f'Вес метрики {weight_name}', fontweight='bold', fontsize=12)
    ax.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()

# ==============================================================================
# 10.3. СРАВНЕНИЕ КАЧЕСТВА АЛГОРИТМОВ
# ==============================================================================

fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Сравнительный анализ производительности алгоритмов', 
             fontsize=16, fontweight='bold')

# График 1: Среднее расстояние
ax1 = axes[0, 0]
algo_names = list(all_algorithms_results.keys())
mean_distances = [np.mean([run['distance'] for run in results['runs']]) 
                   for results in all_algorithms_results.values()]
std_distances = [np.std([run['distance'] for run in results['runs']]) 
                  for results in all_algorithms_results.values()]
colors_list = [colors_algo[name] for name in algo_names]

bars = ax1.bar(algo_names, mean_distances, yerr=std_distances, 
               color=colors_list, alpha=0.7, edgecolor='black', capsize=10)
ax1.set_ylabel('Расстояние Фробениуса', fontsize=11, fontweight='bold')
ax1.set_title('Среднее расстояние до экспертной матрицы', fontweight='bold')
ax1.grid(True, alpha=0.3, axis='y')

for bar, dist, std in zip(bars, mean_distances, std_distances):
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height + std + 0.001,
             f'{dist:.5f}', ha='center', va='bottom', fontweight='bold', fontsize=9)

# График 2: Лучшее расстояние
ax2 = axes[0, 1]
best_distances = [np.min([run['distance'] for run in results['runs']]) 
                   for results in all_algorithms_results.values()]

bars = ax2.bar(algo_names, best_distances, color=colors_list, 
               alpha=0.7, edgecolor='black')
ax2.set_ylabel('Расстояние Фробениуса', fontsize=11, fontweight='bold')
ax2.set_title('Лучшее достигнутое расстояние', fontweight='bold')
ax2.grid(True, alpha=0.3, axis='y')

for bar, dist in zip(bars, best_distances):
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2., height + 0.001,
             f'{dist:.5f}', ha='center', va='bottom', fontweight='bold', fontsize=9)

# График 3: Среднее время выполнения
ax3 = axes[1, 0]
mean_times = [np.mean([run['time'] for run in results['runs']]) 
               for results in all_algorithms_results.values()]
std_times = [np.std([run['time'] for run in results['runs']]) 
              for results in all_algorithms_results.values()]

bars = ax3.bar(algo_names, mean_times, yerr=std_times, 
               color=colors_list, alpha=0.7, edgecolor='black', capsize=10)
ax3.set_ylabel('Время (секунды)', fontsize=11, fontweight='bold')
ax3.set_title('Среднее время выполнения', fontweight='bold')
ax3.grid(True, alpha=0.3, axis='y')

for bar, time_val, std in zip(bars, mean_times, std_times):
    height = bar.get_height()
    ax3.text(bar.get_x() + bar.get_width()/2., height + std + 0.1,
             f'{time_val:.2f}s', ha='center', va='bottom', fontweight='bold', fontsize=9)

# График 4: Стабильность
ax4 = axes[1, 1]
stability = std_distances

bars = ax4.bar(algo_names, stability, color=colors_list, 
               alpha=0.7, edgecolor='black')
ax4.set_ylabel('Стандартное отклонение', fontsize=11, fontweight='bold')
ax4.set_title('Стабильность алгоритма (меньше - лучше)', fontweight='bold')
ax4.grid(True, alpha=0.3, axis='y')

for bar, stab in zip(bars, stability):
    height = bar.get_height()
    ax4.text(bar.get_x() + bar.get_width()/2., height + 0.00001,
             f'{stab:.6f}', ha='center', va='bottom', fontweight='bold', fontsize=9)

plt.tight_layout()
plt.show()

# ==============================================================================
# 10.4. HEATMAP ВЕСОВ ПО АЛГОРИТМАМ
# ==============================================================================

fig, ax = plt.subplots(figsize=(10, 6))

# Создаем матрицу средних весов
weight_matrix = []
for algo_name, results in all_algorithms_results.items():
    weights = np.array([run['weights'] for run in results['runs']])
    mean_weights = weights.mean(axis=0)
    weight_matrix.append(mean_weights)

weight_matrix = np.array(weight_matrix)

# Heatmap
im = ax.imshow(weight_matrix, cmap='YlOrRd', aspect='auto', vmin=0, vmax=1)

# Настройка осей
ax.set_xticks(np.arange(len(weight_names)))
ax.set_yticks(np.arange(len(algo_names)))
ax.set_xticklabels(weight_names, fontsize=11, fontweight='bold')
ax.set_yticklabels(algo_names, fontsize=11, fontweight='bold')

# Добавление значений в ячейки
for i in range(len(algo_names)):
    for j in range(len(weight_names)):
        text = ax.text(j, i, f'{weight_matrix[i, j]:.3f}',
                       ha="center", va="center", color="black", 
                       fontsize=12, fontweight='bold')

ax.set_title('Средние веса метрик кода по алгоритмам', 
             fontsize=14, fontweight='bold', pad=15)
plt.colorbar(im, ax=ax, label='Значение веса')

plt.tight_layout()
plt.show()

# ==============================================================================
# 11. ДЕТАЛЬНЫЙ АНАЛИЗ ЛУЧШЕГО РЕШЕНИЯ
# ==============================================================================

print("\n" + "="*80)
print("ДЕТАЛЬНЫЙ АНАЛИЗ ЛУЧШЕГО РЕШЕНИЯ")
print("="*80)

# Используем лучший результат
optimal_weights = best_run['weights']
best_distance = best_run['distance']

# Создание финальной автоматической матрицы
final_auto_matrix = create_automatic_matrix(optimal_weights)

# Расстояния для отдельных метрик
n = min(final_auto_matrix.shape[0], expert_matrix.shape[0])

dist_halstead = np.linalg.norm(
    halstead_similarity_norm[:n, :n] - expert_matrix[:n, :n], ord='fro'
)
dist_structural = np.linalg.norm(
    structural_similarity_norm[:n, :n] - expert_matrix[:n, :n], ord='fro'
)
dist_token = np.linalg.norm(
    token_similarity_norm[:n, :n] - expert_matrix[:n, :n], ord='fro'
)

print(f"\n📉 Расстояния для отдельных метрик:")
print(f"  Halstead:      {dist_halstead:.6f}")
print(f"  Structural:    {dist_structural:.6f}")
print(f"  Token-based:   {dist_token:.6f}")
print(f"  Комбинация:    {best_distance:.6f}")

# Улучшение относительно лучшей отдельной метрики
best_single = min(dist_halstead, dist_structural, dist_token)
improvement = (best_single - best_distance) / best_single * 100

print(f"\n✨ Улучшение относительно лучшей отдельной метрики: {improvement:.2f}%")

# Статистика различий
differences = []
for i in range(n):
    for j in range(i+1, n):
        differences.append(abs(expert_matrix[i, j] - final_auto_matrix[i, j]))

print(f"\n📊 Статистика различий по всем парам ({len(differences)} пар):")
print(f"  Среднее:             {np.mean(differences):.4f}")
print(f"  Медиана:             {np.median(differences):.4f}")
print(f"  Стандартное откл.:   {np.std(differences):.4f}")
print(f"  Минимум:             {np.min(differences):.4f}")
print(f"  Максимум:            {np.max(differences):.4f}")

# Корреляция Пирсона
flat_expert = expert_matrix[:n, :n].flatten()
flat_auto = final_auto_matrix[:n, :n].flatten()
correlation = np.corrcoef(flat_expert, flat_auto)[0, 1]

print(f"\n🔗 Корреляция Пирсона между матрицами: {correlation:.4f}")

# ==============================================================================
# 12. ВИЗУАЛИЗАЦИЯ ЛУЧШЕГО РЕШЕНИЯ
# ==============================================================================

fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle(f'Анализ лучшего решения ({best_algo_name})', 
             fontsize=16, fontweight='bold')

# 1. Экспертная матрица
im1 = axes[0, 0].imshow(expert_matrix[:n, :n], cmap='RdYlGn', vmin=0, vmax=1)
axes[0, 0].set_title('Экспертная матрица', fontweight='bold')
axes[0, 0].set_xlabel('Фрагмент кода')
axes[0, 0].set_ylabel('Фрагмент кода')
plt.colorbar(im1, ax=axes[0, 0])

# 2. Автоматическая матрица
im2 = axes[0, 1].imshow(final_auto_matrix[:n, :n], cmap='RdYlGn', vmin=0, vmax=1)
axes[0, 1].set_title('Автоматическая матрица (оптимизированная)', fontweight='bold')
axes[0, 1].set_xlabel('Фрагмент кода')
axes[0, 1].set_ylabel('Фрагмент кода')
plt.colorbar(im2, ax=axes[0, 1])

# 3. Разница между матрицами
difference_matrix = np.abs(expert_matrix[:n, :n] - final_auto_matrix[:n, :n])
im3 = axes[1, 0].imshow(difference_matrix, cmap='Reds', vmin=0, vmax=1)
axes[1, 0].set_title('Абсолютная разница', fontweight='bold')
axes[1, 0].set_xlabel('Фрагмент кода')
axes[1, 0].set_ylabel('Фрагмент кода')
plt.colorbar(im3, ax=axes[1, 0])

# 4. Scatter plot корреляции
mask = np.triu(np.ones((n, n), dtype=bool), k=1)
expert_values = expert_matrix[:n, :n][mask]
auto_values = final_auto_matrix[:n, :n][mask]

axes[1, 1].scatter(expert_values, auto_values, alpha=0.5, s=50, 
                   color=colors_algo[best_algo_name])
axes[1, 1].plot([0, 1], [0, 1], 'r--', linewidth=2, label='Идеальное соответствие')
axes[1, 1].set_xlabel('Экспертная оценка', fontsize=11, fontweight='bold')
axes[1, 1].set_ylabel('Автоматическая оценка', fontsize=11, fontweight='bold')
axes[1, 1].set_title(f'Корреляция: {correlation:.4f}', fontweight='bold')
axes[1, 1].legend(fontsize=10)
axes[1, 1].grid(True, alpha=0.3)
axes[1, 1].set_xlim([0, 1])
axes[1, 1].set_ylim([0, 1])

plt.tight_layout()
plt.show()

# ==============================================================================
# 13. СОХРАНЕНИЕ РЕЗУЛЬТАТОВ
# ==============================================================================

print("\n" + "="*80)
print("СОХРАНЕНИЕ РЕЗУЛЬТАТОВ")
print("="*80)

# Сравнительная таблица алгоритмов
comparison_df.to_csv('algorithms_comparison_code.csv', index=False)
print("✓ Сравнение алгоритмов → algorithms_comparison_code.csv")

# Оптимальные веса лучшего решения
weights_df = pd.DataFrame({
    'Метрика': ['Halstead', 'Structural', 'Token-based'],
    'Оптимальный_вес': optimal_weights,
    'Процент': optimal_weights * 100
})
weights_df.to_csv('optimal_weights_code_best.csv', index=False)
print("✓ Оптимальные веса лучшего решения → optimal_weights_code_best.csv")

# Детальные результаты всех запусков всех алгоритмов
all_runs_detailed = []
for algo_name, results in all_algorithms_results.items():
    for run in results['runs']:
        all_runs_detailed.append({
            'Алгоритм': algo_name,
            'Запуск': run['run_id'],
            'Вес_Halstead': run['weights'][0],
            'Вес_Structural': run['weights'][1],
            'Вес_Token': run['weights'][2],
            'Расстояние': run['distance'],
            'Время_(с)': run['time']
        })

all_runs_df = pd.DataFrame(all_runs_detailed)
all_runs_df.to_csv('all_algorithms_detailed_runs_code.csv', index=False)
print("✓ Детальные результаты всех запусков → all_algorithms_detailed_runs_code.csv")

# Автоматическая матрица
auto_matrix_df = pd.DataFrame(
    final_auto_matrix[:n, :n],
    index=[f"Код_{i+1}" for i in range(n)],
    columns=[f"Код_{i+1}" for i in range(n)]
)
auto_matrix_df.to_csv('automatic_similarity_matrix_code_best.csv')
print("✓ Автоматическая матрица лучшего решения → automatic_similarity_matrix_code_best.csv")

# Сводная статистика
summary_data = {
    'Метрика': [
        'Лучший алгоритм',
        'Лучшее расстояние',
        'Среднее расстояние лучшего алгоритма',
        'Расстояние Halstead',
        'Расстояние Structural',
        'Расстояние Token-based',
        'Улучшение (%)',
        'Корреляция Пирсона',
        'Средняя разница',
        'Медиана разницы',
        'Оптимальный вес Halstead',
        'Оптимальный вес Structural',
        'Оптимальный вес Token-based'
    ],
    'Значение': [
        best_algo_name,
        best_distance,
        np.mean([run['distance'] for run in best_algo_results['runs']]),
        dist_halstead,
        dist_structural,
        dist_token,
        improvement,
        correlation,
        np.mean(differences),
        np.median(differences),
        optimal_weights[0],
        optimal_weights[1],
        optimal_weights[2]
    ]
}

summary_df = pd.DataFrame(summary_data)
summary_df.to_csv('optimization_summary_code_final.csv', index=False)
print("✓ Финальная сводная статистика → optimization_summary_code_final.csv")

print("\n" + "="*80)
print("ОПТИМИЗАЦИЯ ЗАВЕРШЕНА УСПЕШНО!")
print("="*80)
print(f"\n🎯 Финальные результаты (лучший алгоритм: {best_algo_name}):")
print(f"   • Оптимальный вес Halstead:    {optimal_weights[0]:.4f} ({optimal_weights[0]*100:.1f}%)")
print(f"   • Оптимальный вес Structural:  {optimal_weights[1]:.4f} ({optimal_weights[1]*100:.1f}%)")
print(f"   • Оптимальный вес Token-based: {optimal_weights[2]:.4f} ({optimal_weights[2]*100:.1f}%)")
print(f"   • Расстояние Фробениуса:       {best_distance:.6f}")
print(f"   • Корреляция Пирсона:          {correlation:.4f}")
print(f"   • Улучшение:                   {improvement:.2f}%")
```

Выводы
1. Метрики Хэлстеда оказались неинформативны для оценки сходства программного кода в данной задаче - оптимальный вес составил 0.0% для всех успешных алгоритмов. Это критически важное открытие, которое показывает, что сложностные метрики Хэлстеда (словарь, объем, усилия) не коррелируют с экспертным восприятием сходства кода.
2. Структурные метрики являются доминирующими с весом 63.2%. Это означает, что эксперты в первую очередь оценивают сходство кода по структурным характеристикам: количеству строк, глубине вложенности, количеству функций, циклов, условий и общей архитектуре кода.
3. Token-based метрики занимают второе место с весом 36.8%, что показывает важность лексического сходства - совпадения идентификаторов, структуры токенов и используемых конструкций языка.
4. DE является абсолютным лидером для оптимизации метрик кода - самый быстрый (0.39 с), наиболее стабильный (разброс 10⁻¹⁴) и достигает глобального оптимума во всех 10 запусках. Для данного класса задач это оптимальный выбор.
5. PSO показывает сопоставимое качество с DE при немного большем времени (0.54 с) и минимально большем разбросе (10⁻⁷), что также делает его отличным выбором.
6. GA демонстрирует принципиальную неспособность найти глобальный оптимум - средний результат 1.925944 против 1.924730 у лидеров, огромная вариация весов (CV до 103% для Хэлстеда) и нестабильность результатов (стандартное отклонение 0.0013). Это делает GA непригодным для данной задачи.
7. Качество аппроксимации умеренное - корреляция 0.7605 ниже, чем для текстовых документов (0.8012), что говорит о том, что автоматическая оценка сходства кода сложнее, чем оценка сходства текстов.
8. Комбинированный подход дает улучшение 19.41% по сравнению с лучшей отдельной метрикой (структурные метрики). Это значительное, но меньшее улучшение, чем для текстов (28.59%), что может указывать на то, что для кода структурные метрики уже хорошо улавливают сходство.
9. Средняя ошибка 0.15 при диапазоне [0, 1] означает относительную погрешность около 15%, что сопоставимо с результатами для текстов (14%). Медиана ошибки (0.12) ниже среднего, указывая на наличие отдельных выбросов.
10. Критическая находка исследования: Метрики Хэлстеда, несмотря на их широкое использование в оценке сложности программного обеспечения, не коррелируют с экспертной оценкой сходства кода. Это требует пересмотра подходов к автоматической оценке сходства программ и указывает на необходимость фокусировки на структурных и лексических характеристиках.
11. Рекомендация для практического применения: Использовать DE (дифференциальную эволюцию) с комбинацией структурных метрик (63%) и token-based метрик (37%), полностью исключив метрики Хэлстеда из рассмотрения. Это обеспечит оптимальное соотношение точности, скорости и стабильности результатов.
